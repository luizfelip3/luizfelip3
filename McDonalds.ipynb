```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/mc_donalds.csv')
df.head()
```

```python
# Verificar se tem nulo
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Verificar se tem dados duplicados
duplicados = df[df.duplicated(keep='first')]
print(f'{duplicados}')
```

```python
# Retornar o valor maximo de cada coluna
max = df.max(axis=0)
print(max)
```

```python
# Retonar qual o lanche que tem o maior Sódio
sodium = df['Sodium (mg)']==2399.49
maxSodium = df[sodium]
maxSodium.head()
```

```python
energy = df['Energy (kCal)']==834.36
maxEnergy = df[energy]
maxEnergy.head()
```

```python
# Criando um grafico de barra para mostrar o item com maior Sódio
plt.figsize=(20,20)
pltSodio = 2399.49
pltSodioMax = 2000
plt.title('Ghee Rice with Mc Spicy Fried Chicken') # Titulo
plt.barh(pltSodioMax,pltSodio, color='red') #Gerando o grafico
plt.ylabel('Sódio Maximo = 2000 (mg)')
plt.xlabel('Sódio do item')
```
